```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SVM:

    def __init__(self, learning_rate : float = 0.01,
                 lamda_param : float = 0.01, n_iters : int = 10000) -> None:
        self.learning_rate = learning_rate
        self.lamda_param = lamda_param
        self.n_iters = n_iters
        self.weights = None
        self.bias = None


    def fit(self, X : np.ndarray, y : np.ndarray) -> None:
        n_samples, n_features = X.shape
        y_transformed = np.where(y <= 0, -1, 1)

        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            for idx, point in enumerate(X):
                if y_transformed[idx] * (np.dot(point, self.weights) - self.bias) >= 1:
                    self.weights -= self.learning_rate * (2 * self.lamda_param * self.weights)
                else:
                    self.weights -= self.learning_rate * (
                        2 * self.lamda_param * self.weights - np.dot(point, y_transformed[idx]))
                    self.bias -= self.learning_rate * y_transformed[idx]


    def predict(self, X : np.ndarray) -> np.ndarray:
        predictions = np.dot(X, self.weights) - self.bias
        return np.sign(predictions)
```
